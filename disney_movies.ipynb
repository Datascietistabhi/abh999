```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sn
```

```python
df=pd.read_csv("a.csv")
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.fillna(np.NaN,inplace=True)
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
df['year'] = df['release_date'].dt.date.astype(str).str.split('-', expand=True)[0]
```

```python
df.head()
```

```python
df.sort_values('total_gross',ascending=False).head(5)
```

```python
plt.style.use('dark_background')
```

```python
sn.countplot(y='genre',data=df)
```

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=90)
sn.histplot(df['year'],kde=True)
```

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=90)
sn.histplot(x='year',hue='genre',data=df)
```

```python
sn.barplot(y='genre',x='total_gross',data=df)
```

```python
plt.figure(figsize=(12,6))
sn.boxplot(y='genre',x='inflation_adjusted_gross',data=df)
```

```python
sn.countplot(x='mpaa_rating',data=df)
```

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=90)
sn.lineplot(x='year', y='inflation_adjusted_gross', data=df)
```

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=90)
sn.lineplot(x='year', y='total_gross', data=df)
```

```python
plt.figure(figsize=(8,4))
sn.countplot(x='mpaa_rating',hue='genre',data=df)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
